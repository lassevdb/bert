```python
import torch
import torch.nn.functional as F
from transformers import AutoModelForMaskedLM, AutoTokenizer, utils
import re
from datasets import load_dataset
from bertviz import model_view, head_view
from bertviz.transformers_neuron_view import BertModel, BertTokenizer
from bertviz.neuron_view import show
utils.logging.set_verbosity_error()
```

```python
from bertviz import head_view, model_view
from transformers import BertTokenizer, BertModel

model_version = 'dbmdz/bert-base-italian-cased'
model = BertModel.from_pretrained(model_version, output_attentions=True)
tokenizer = BertTokenizer.from_pretrained(model_version)

sentence_a = "È possibile che, anche se è strano, sia felice."
inputs = tokenizer.encode_plus(sentence_a, return_tensors='pt')
input_ids = inputs['input_ids']
token_type_ids = inputs['token_type_ids']
attention = model(input_ids, token_type_ids=token_type_ids)[-1]
input_id_list = input_ids[0].tolist() # Batch index 0
tokens = tokenizer.convert_ids_to_tokens(input_id_list) 
```

```python
model_view(attention, tokens, display_mode="light")
```
